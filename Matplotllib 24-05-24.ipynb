```python
x = a['Name']
y = a['Marks']
```

```python
#plt.plot(x,y, marker='o') marker color,line styles
#plt.plot(x,y,'<:m')
#plt.plot(x,y,'*-.')
#plt.plot(x,y,'*')
plt.plot(x,y,'*m', ms=15, lw=5) #increase marker line size , lw= increase line width
plt.xlabel('Marks')
plt.ylabel('Name')
plt.title('ADIT', {'color':'red' , 'size':12, 'family':'serif'})
plt.grid(visible=None, axis='x' ,color='green', linestyle='-.')
```
