```python
import pandas as pd 
from sklearn import linear_model
```

```python
df = pd.read_csv('houseprice.csv')
df.head()
```

```python
df.info()
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(df.drop('price', axis='columns'),df.price)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[800,2,10]])
```
