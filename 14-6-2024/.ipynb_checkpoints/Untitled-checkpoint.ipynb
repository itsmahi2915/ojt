```python
import pandas as pd 
```

```python
df = pd.read_csv('score1.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('price',axis=1),df.price)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
df1 = pd.DataFrame({'original value':'y_test','prediction':y_pred})
```

```python
df1.to_csv('mob.csv')
```

```python
from sklearn.metrics import mean_absolute_error as mae
mae_error = mae(y_test,y_pred)  #mae = mean absolute error
mae_error
```

```python
y_pred
```

```python
from sklearn.metrics import  mean_squared_error as mse 
mse(y_test,y_pred)
```
