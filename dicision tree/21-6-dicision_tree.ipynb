```python
from sklearn import datasets
```

```python
df = datasets.load_iris()
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df.data
y = df.target
```

```python
x
```

```python
y
```

```python
df.feature_names
```

```python
df.target_names
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
from sklearn import tree
text = tree.export_text(model)
text
```

```python
from sklearn import tree
text = tree.export_text(model,feature_names=df.feature_names, class_names=df.target_names)
text
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,20))
a = tree.plot_tree(model,feature_names=df.feature_names, class_names=df.target_names, filled=True)
```
