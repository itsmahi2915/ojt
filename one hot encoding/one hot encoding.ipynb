```python
import pandas as pd 
```

```python
#dummies
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
df_dummies = pd.concat([df, dummies],axis='columns')
df_dummies
```

```python
df_dummies.drop(['town','delhi'],axis='columns',inplace=True)
df_dummies
```

```python
df_dummies.replace([True,False],[1,0], inplace=True)
df_dummies
```

```python
x = df_dummies.drop('price',axis='columns')
x
```

```python
y = df_dummies.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[2600,0,1]])
```

```python
model.predict([[2600,1,0]])
```
